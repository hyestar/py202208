```python
import matplotlib
import matplotlib.font_manager as fm
# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout=True)
# 폰트지정
font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font',family = fontname)
a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)
x = range(0,10)
y = range(5,15)
a1.plot(x,y, "m^")
x = range(0,50)
y = range(0,50)
a2.plot(x,y,"ro")
# x축, y축이름 지정
a1.set_xlabel('시간')
a1.set_ylabel('돈')
a2.set_xlabel("시간")
a2.set_ylabel('돈')
plt.show()
```

```python
name = ['홍길동', '홍길순', '임꺽정']
age = [23, 22, 50]
home = ['대전', '서울', '부산']
member3 = pd.DataFrame()
member3['이름'] = name
member3['나이'] = age
member3['집'] = home
# 데이터프레임에 성별 컬럼을 추가
gender = pd.Series(['남자', '여자', '남자'], name="성별")
member3["성별"] = gender
print(member3)
# 집계함수
# 최솟값
print(member3["나이"].min())
# 최대값
print(member3["나이"].max())
# 평균값
print(member3["나이"].mean())
# 데이터 프레임의 기본 통계 내용을 출력
print(member3.describe())
# head(끊어오려는 행의 수) => 위에서부터 데이터를 끊어옴
print(member3.head(2))
# tail(끊어오려는 행의 수) => 밑에서부터 데이터를 끊어옴
print(member3.tail(1))
```
