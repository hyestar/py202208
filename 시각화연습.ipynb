```python
fig = plt.figure()
a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)
x = range(0,10)
y = range(5,15)
a1.plot(x,y, "m^")
x = range(0,50)
y = range(0,50)
a2.plot(x,y,"ro")
plt.show()
```

```python
import matplotlib
# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout=True)
a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)
x = range(0,10)
y = range(5,15)
a1.plot(x,y, "m^")
x = range(0,50)
y = range(0,50)
a2.plot(x,y,"ro")
# x축, y축이름 지정
a1.set_xlabel('time')
a1.set_ylabel('money')
a2.set_xlabel("time")
a2.set_ylabel('money')
plt.show()
```

```python

fig = plt.figure()
plt.style.use('ggplot')
a = fig.add_subplot(111)
x = [3,4]
y = [1,2]
labels = ['english', 'math']
colors = ['red', 'blue']
a.bar(x,y, tick_label = labels, color = colors)
plt.show()
```

```python
fig = plt.figure()
plt.style.use('ggplot')
a = fig.add_subplot(111)
x = [3,4]
y = [1,2]
labels = ['english', 'math']
colors = ['red', 'blue']
a.barh(x,y, tick_label = labels, color = colors)
plt.show()
```

```python
# 막대그래프에 수치 표현
fig = plt.figure()
plt.style.use('ggplot')
a = fig.add_subplot(111)
x = [3,4]
y = [1,2]
labels = ['english', 'math']
colors = ['red', 'blue']
a.bar(x,y, tick_label = labels, color = colors)
for i, v in enumerate(y):
    a.text(v+2, i+1.05, str(v))
plt.show()
```
